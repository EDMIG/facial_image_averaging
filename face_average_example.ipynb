```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

# Presidents example

```python
from face_average import readPoints, readImages, face_avg
```

```python
facial_landmarks = readPoints('us_presidents')
images = readImages('us_presidents')
```

facial_landmarks text files ar generated for every singele image by face_landmarks.py before averaging. It first detect face with pretrained model and then using dlib algorithm generate facial landmarks

```python
print(facial_landmarks[0])
```

These point now can be used for detecting eyes for example and centering faces in same position. For example plot 2 images.

```python
for i, img in enumerate(images[0:2]):

    # cv2.drawContours(images[0], facial_landmarks[0], -1, (0,255,0), 3)
    plt.figure(figsize=(20,10))
    
    img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    for landmark in facial_landmarks[i]:
        cv2.circle(img_bgr, landmark, 10, (0,255,255), -1)
    
    plt.imshow(img_bgr)
```
